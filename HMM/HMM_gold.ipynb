```python
#Cell 1:Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from hmmlearn import hmm
from sklearn.preprocessing import StandardScaler
import warnings
from datetime import datetime, timedelta
from scipy import stats
import matplotlib.dates as mdates
from sklearn.metrics import silhouette_score
from sklearn.model_selection import train_test_split
import pickle

warnings.filterwarnings('ignore')

plt.style.use('fivethirtyeight')
sns.set(font_scale=1.2)
plt.rcParams['figure.figsize'] = (14, 8)

#Cell 2:Load the data from CSV
# Ensure this path points to your data file
csv_file_path = r"C:\Users\User\Downloads\KPZHackathon\merged_btc_gold_hourly_data_filtered.csv"

print(f"Reading data from CSV file: {csv_file_path}")

#Cell 3:Process the CSV data (Modified for robust datetime parsing)
try:
    df = pd.read_csv(csv_file_path)
    
    print("Successfully loaded CSV data")
    print("Data shape:", df.shape)
    print("Columns:", df.columns.tolist())
    
    # Ensure 'datetime' column is in datetime format
    if 'datetime' in df.columns:
        df['datetime'] = pd.to_datetime(df['datetime'])
        print("Converted 'datetime' column to datetime objects.")
    elif 'ts' in df.columns: # [cite: 2]
        df['datetime'] = pd.to_datetime(df['ts'], unit='ms') # [cite: 2]
        print("Converted 'ts' column to datetime objects and named it 'datetime'.")
    else:
        print("Warning: Neither 'datetime' nor 'ts' column found for time series analysis. Ensure your CSV has a recognizable date column.")
        # Handle error or proceed with caution if datetime is not critical for some parts
   
    print("First few rows:")
    print(df.head())
    
except Exception as e:
    print(f"Error loading CSV file: {e}")

   
# Cell 4: Display data statistics
print("\nBasic statistics:")
print(df.describe())

#Cell 5:Features (MODIFIED to include BTC Close and Gold Close)
# Explicitly define the features for HMM analysis
# These should match column names in your CSV:
# 'flow_mean', 'flow_total' (from original script's likely intent)
# 'transactions_count_flow_1' (assuming this is the primary transaction count)
# 'Close' (BTC closing price)
# 'gold_Close' (Gold closing price)

features = ['flow_mean', 'flow_total', 'transactions_count_flow_1', 'Close', 'gold_Close']

# Verify that all specified features exist in the DataFrame
missing_features = [f for f in features if f not in df.columns]
if missing_features:
    print(f"Critical Error: The following specified features are MISSING from your CSV file: {missing_features}")
    print(f"Available columns in your CSV are: {df.columns.tolist()}")
    print("Please check the 'features' list in Cell 5 and your CSV column names.")
    # You might want to raise an error here or exit if features are missing
    raise ValueError(f"Missing features required for HMM analysis: {missing_features}")
else:
    print(f"\nUsing the following features for HMM analysis: {features}")

#Cell 6:Prepare data for analysis
df = df.sort_values('datetime').reset_index(drop=True)
print("\nMissing values in each column before fill:")
print(df[features].isnull().sum()) # Check missing values only in selected features

# Fill missing values for selected features
df[features] = df[features].fillna(method='ffill')
df[features] = df[features].fillna(method='bfill') # Backfill any remaining NaNs at the beginning

print("\nMissing values in selected features after fill:")
print(df[features].isnull().sum())

# Check if any NaNs remain after ffill and bfill in selected features
if df[features].isnull().any().any():
    print("Warning: NaNs still present in feature columns after forward and backward fill. Consider data imputation or review source data.")
    print(df[features][df[features].isnull().any(axis=1)])


#Cell 7:Plot time series data
fig, axes = plt.subplots(len(features), 1, figsize=(14, 4*len(features)), sharex=True)

if len(features) == 1:
    axes = [axes] # type: ignore

for i, feature in enumerate(features):
    axes[i].plot(df['datetime'], df[feature], label=feature)
    axes[i].set_title(f'{feature} Over Time')
    axes[i].set_ylabel(feature)
    axes[i].legend()
    axes[i].grid(True)

plt.xlabel('Date')
plt.tight_layout()
plt.show()

#Cell 8:Prepare features and standardize
X = df[features].copy()

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_df = pd.DataFrame(X_scaled, columns=features, index=df.index)

#Cell 9:Plot feature correlations
plt.figure(figsize=(10, 8))
sns.heatmap(X.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Feature Correlation Matrix')
plt.tight_layout()
plt.show()

#Cell 10:Plot feature distributions
fig, axes = plt.subplots(len(features), 1, figsize=(14, 4*len(features)), sharex=False)
if len(features) == 1:
    axes = [axes] # type: ignore

for i, feature in enumerate(features):
    sns.histplot(X[feature], kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {feature}')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

#Cell 11:Define function to find optimal number of states
def find_optimal_states(X_scaled_data, max_states=10):
    n_states_range = range(2, max_states + 1)
    bic_scores = []
    aic_scores = []
    silhouette_scores = []
    
    print("Finding optimal number of states (this may take some time)...")
    for n_states in n_states_range:
        print(f"Testing model with {n_states} states...")
        model_opt = hmm.GaussianHMM(n_components=n_states, covariance_type='full', 
                                n_iter=100, random_state=42, tol=0.01) # Added tol for convergence
        try:
            model_opt.fit(X_scaled_data)
            
            n_features = X_scaled_data.shape[1]
            # Parameters:
            # Transition matrix: n_states * (n_states - 1) (since rows sum to 1)
            # Initial state probabilities: n_states - 1 (since they sum to 1)
            # Emission means: n_states * n_features
            # Emission covariances: n_states * n_features * (n_features + 1) / 2 (for full covariance)
            n_params = (n_states * (n_states - 1)) + \
                       (n_states - 1) + \
                       (n_states * n_features) + \
                       (n_states * n_features * (n_features + 1) / 2)

            log_likelihood = model_opt.score(X_scaled_data)
            bic = -2 * log_likelihood + n_params * np.log(X_scaled_data.shape[0])
            aic = -2 * log_likelihood + 2 * n_params
            
            bic_scores.append(bic)
            aic_scores.append(aic)
            hidden_states_opt = model_opt.predict(X_scaled_data)
            
            if len(np.unique(hidden_states_opt)) > 1: # Silhouette score requires at least 2 labels
                try:
                    silhouette = silhouette_score(X_scaled_data, hidden_states_opt)
                    silhouette_scores.append(silhouette)
                except ValueError:
                    print(f"Could not compute silhouette score for {n_states} states (possibly single cluster predicted). Appending NaN.")
                    silhouette_scores.append(np.nan)
            else:
                silhouette_scores.append(np.nan)
        except Exception as e_fit:
            print(f"Error fitting HMM with {n_states} states: {e_fit}. Appending NaN to scores.")
            bic_scores.append(np.nan)
            aic_scores.append(np.nan)
            silhouette_scores.append(np.nan)

    plt.figure(figsize=(12, 8))
    plt.plot(n_states_range, bic_scores, 'o-', label='BIC')
    plt.plot(n_states_range, aic_scores, 's-', label='AIC')
    plt.xlabel('Number of States')
    plt.ylabel('Score (Lower is Better)')
    plt.title('BIC and AIC Scores for Different Numbers of States')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    plt.figure(figsize=(12, 8))
    plt.plot(n_states_range, silhouette_scores, 'o-', label='Silhouette Score (Higher is Better)')
    plt.xlabel('Number of States')
    plt.ylabel('Silhouette Score')
    plt.title('Silhouette Scores for Different Numbers of States')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Find optimal n_states (handle NaNs in scores)
    valid_bic_scores = np.array(bic_scores)[~np.isnan(bic_scores)]
    valid_n_states_bic = np.array(n_states_range)[~np.isnan(bic_scores)]
    
    if len(valid_bic_scores) > 0:
        optimal_n_states_bic = valid_n_states_bic[np.argmin(valid_bic_scores)]
        print(f"Optimal number of states based on BIC: {optimal_n_states_bic}")
    else:
        print("Could not determine optimal states based on BIC due to errors.")
        optimal_n_states_bic = 3 # Default or ask user
        print(f"Defaulting to {optimal_n_states_bic} states.")

    # You might also consider AIC or Silhouette Score for choosing optimal_n_states
    # For now, we use BIC as per the original script's implication.
    return optimal_n_states_bic

#Cell 12:Determine and set optimal number of states
# Uncomment the line below to dynamically find optimal states.
# This can be time-consuming.
# optimal_n_states = find_optimal_states(X_scaled, max_states=8) # Reduced max_states for quicker run if uncommented

# For consistency with the original script, or if you have a preferred number:
print("Attempting to find optimal number of states for the HMM...")
optimal_n_states = 8


#Cell 13:HMM model
model = hmm.GaussianHMM(n_components=optimal_n_states, 
                        covariance_type='full', 
                        n_iter=1000, # Increased iterations
                        random_state=42,
                        tol=0.001) # Added tolerance

print("Fitting HMM model...")
model.fit(X_scaled)
print("HMM model fitting complete.")

hidden_states = model.predict(X_scaled)
df['regime'] = hidden_states

#Cell 14:Visualize the identified regimes
plt.figure(figsize=(16, 10))
colors = plt.cm.get_cmap('tab10', optimal_n_states) # More robust way to get colors

# Plotting the first feature with regimes
plot_feature_for_regime_viz = features[0] 

for state in range(optimal_n_states):
    mask = (df['regime'] == state)
    if np.any(mask): # Only plot if the state is present
        plt.scatter(df.loc[mask, 'datetime'], 
                    df.loc[mask, plot_feature_for_regime_viz], 
                    c=[colors(state)], # Use colormap directly
                    label=f'Regime {state}',
                    alpha=0.7,
                    s=30)

plt.title(f'Data with Identified Regimes ({plot_feature_for_regime_viz})')
plt.xlabel('Date')
plt.ylabel(plot_feature_for_regime_viz)
plt.legend()
plt.grid(True)
plt.show()

#Cell 15:Calculate regime statistics
regime_stats = df.groupby('regime')[features].agg(['mean', 'std', 'min', 'max'])
print("\nRegime statistics:")
print(regime_stats)

#Cell 16:Save the DataFrame with regimes to a CSV file
output_csv_path = r"C:\Users\User\Downloads\output_data_with_regimes_btc_gold.csv" # Modified output file name
try:
    df.to_csv(output_csv_path, index=False)
    print(f"\nDataFrame with regimes successfully saved to: {output_csv_path}")
except Exception as e:
    print(f"\nError saving DataFrame to CSV: {e}")

#Cell 16 (should be 17):Plot regime characteristics
fig, axes = plt.subplots(len(features), 1, figsize=(14, 4*len(features)), sharex=True)
if len(features) == 1:
    axes = [axes] # type: ignore

for i, feature in enumerate(features):
    means = [regime_stats.loc[state, (feature, 'mean')] if state in regime_stats.index else np.nan for state in range(optimal_n_states)]
    stds = [regime_stats.loc[state, (feature, 'std')] if state in regime_stats.index else np.nan for state in range(optimal_n_states)]
    
    axes[i].bar(range(optimal_n_states), means, yerr=stds, capsize=10, 
                color=[colors(j) for j in range(optimal_n_states)], alpha=0.7)
    axes[i].set_title(f'Mean {feature} by Regime with Standard Deviation')
    axes[i].set_ylabel(feature)
    axes[i].set_xticks(range(optimal_n_states))
    axes[i].set_xticklabels([f'Regime {state}' for state in range(optimal_n_states)])
    axes[i].grid(True, axis='y')

plt.xlabel('Regime') # Add x-label to the bottom plot
plt.tight_layout()
plt.show()

#Cell 17 (should be 18):Analyze regime transitions
# Ensure all regimes are present in transitions table, fill with 0 if not
all_regimes = list(range(optimal_n_states))
transitions = pd.crosstab(df['regime'].shift(1), df['regime'], 
                         rownames=['From'], colnames=['To'])
transitions = transitions.reindex(index=all_regimes, columns=all_regimes, fill_value=0)

print("\nRegime transition counts:")
print(transitions)

plt.figure(figsize=(10, 8))
sns.heatmap(transitions, annot=True, cmap='Blues', fmt='d')
plt.title('Regime Transition Counts')
plt.tight_layout()
plt.show()

#Cell 18 (should be 19):Calculate transition probabilities
transition_probs = transitions.div(transitions.sum(axis=1), axis=0).fillna(0) # Fill NaN if a state never transitions out
print("\nRegime transition probabilities:")
print(transition_probs)

plt.figure(figsize=(10, 8))
sns.heatmap(transition_probs, annot=True, cmap='Blues', fmt='.2f')
plt.title('Regime Transition Probabilities')
plt.tight_layout()
plt.show()

#Cell 19 (should be 20):Analyze regime durations
df['regime_shifted'] = df['regime'].shift(1)
df['regime_changed'] = (df['regime'] != df['regime_shifted'])
# A new regime starts at index 0, or when regime_changed is True
df['regime_start'] = df['regime_changed'].fillna(True) # First row is a start
df.loc[0, 'regime_start'] = True


regime_group_ids = df['regime_start'].cumsum()
df['regime_group_id'] = regime_group_ids


durations_list = []
for group_id, group_df in df.groupby('regime_group_id'):
    if not group_df.empty:
        regime = group_df['regime'].iloc[0]
        start_date = group_df['datetime'].iloc[0]
        end_date = group_df['datetime'].iloc[-1]
        # Duration in hours, then convert to days if needed
        duration_hours = (end_date - start_date).total_seconds() / 3600.0 + 1 # +1 for hourly data, counting the start hour
        duration_days = duration_hours / 24.0
        
        durations_list.append({
            'regime': regime,
            'start_date': start_date,
            'end_date': end_date,
            'duration_hours': duration_hours,
            'duration_days': duration_days
        })

duration_df = pd.DataFrame(durations_list)

print("\nRegime duration statistics (in hours):")
print(duration_df.groupby('regime')['duration_hours'].agg(['count', 'mean', 'min', 'max', 'std']))


#Cell 20 (should be 21):Plot regime durations (in hours)
plt.figure(figsize=(12, 6))
for state in range(optimal_n_states):
    state_durations = duration_df[duration_df['regime'] == state]['duration_hours']
    if not state_durations.empty:
        plt.hist(state_durations, alpha=0.7, 
                 label=f'Regime {state} (mean: {state_durations.mean():.2f} hrs)', 
                 bins=20, color=colors(state)) # type: ignore

plt.title('Distribution of Regime Durations (in Hours)')
plt.xlabel('Duration (hours)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()


#Cell 21 (should be 22):Train/test split for validation
train_size = int(len(X_scaled) * 0.8)
X_train, X_test = X_scaled[:train_size], X_scaled[train_size:]
dates_train, dates_test = df['datetime'][:train_size], df['datetime'][train_size:]
df_train, df_test = df.iloc[:train_size].copy(), df.iloc[train_size:].copy()


train_model = hmm.GaussianHMM(n_components=optimal_n_states, 
                            covariance_type='full', 
                            n_iter=1000, 
                            random_state=42,
                            tol=0.001)
print("\nFitting HMM model on training data...")
train_model.fit(X_train)
print("Training HMM model fit complete.")

train_regimes = train_model.predict(X_train)
test_regimes = train_model.predict(X_test)

df_train['regime_predicted'] = train_regimes
df_test['regime_predicted'] = test_regimes


#Cell 22 (should be 23):Visualize training and test regimes
plt.figure(figsize=(16, 8))
plot_feature_split_viz = features[0]

plt.plot(df['datetime'], df[plot_feature_split_viz], 'k-', alpha=0.3, label=f'Full Data {plot_feature_split_viz}')
plt.axvline(x=dates_train.iloc[-1], color='r', linestyle='--', label='Train-Test Split')

# Plot True Regimes (from full model) for comparison context
# for state in range(optimal_n_states):
#     mask_true_train = (df_train['regime'] == state)
#     if np.any(mask_true_train):
#         plt.scatter(dates_train[mask_true_train],
#                     df_train.loc[mask_true_train, plot_feature_split_viz],
#                     marker='o', s=80, facecolors='none', edgecolors=[colors(state)], # type: ignore
#                     label=f'True Train Regime {state}' if not plt.gca().get_legend_handles_labels()[1] else None, # Avoid duplicate labels
#                     alpha=0.5)

# Plot Predicted Regimes on Train Data
for state in range(optimal_n_states):
    mask_train = (train_regimes == state)
    if np.any(mask_train):
        plt.scatter(dates_train[mask_train], 
                    df_train.loc[mask_train, plot_feature_split_viz], 
                    c=[colors(state)], # type: ignore
                    label=f'Predicted Train Regime {state}',
                    alpha=0.7, s=30)

# Plot Predicted Regimes on Test Data
for state in range(optimal_n_states):
    mask_test = (test_regimes == state)
    if np.any(mask_test):
        plt.scatter(dates_test[mask_test], 
                    df_test.loc[mask_test, plot_feature_split_viz], 
                    c=[colors(state)], # type: ignore
                    marker='x',
                    label=f'Predicted Test Regime {state}',
                    alpha=0.7, s=50)


plt.title(f'Data Regimes: Training vs. Testing Predictions ({plot_feature_split_viz})')
plt.xlabel('Date')
plt.ylabel(plot_feature_split_viz)
plt.legend(loc='upper left', bbox_to_anchor=(1,1)) # Adjust legend position
plt.grid(True)
plt.tight_layout(rect=[0, 0, 0.85, 1]) # Adjust layout for legend
plt.show()

#Cell 23 (should be 24): Prediction and simulation (Using the model trained on full data: `model`)

def simulate_future_regimes(hmm_model, current_observation_scaled, n_steps=30, random_seed=42):
    """ Simulates future states and observations from an HMM. """
    current_obs = current_observation_scaled.copy()
    sim_states = []
    sim_observations_scaled = []
    
    # Use model's random state if available, otherwise create one
    # hmmlearn's predict and sample methods handle randomness internally based on model's random_state
    # For reproducibility of the simulation path IF model.sample were to be used, we'd need a seed.
    # model.predict is deterministic given the model and observation.
    # For generating next_obs from means/covars, we need a random state.
    
    local_random_state = np.random.RandomState(random_seed)

    # Get current state based on last observation
    # Reshape if current_obs is 1D
    if current_obs.ndim == 1:
        current_obs_reshaped = current_obs.reshape(1, -1)
    else:
        current_obs_reshaped = current_obs

    current_state = hmm_model.predict(current_obs_reshaped)[0]

    for _ in range(n_steps):
        sim_states.append(current_state) # Store current state before transition

        # Transition to next state based on transition probabilities
        # hmm_model.transmat_ is (n_components, n_components)
        next_state_probs = hmm_model.transmat_[current_state, :]
        next_state = local_random_state.choice(hmm_model.n_components, p=next_state_probs)
        
        # Generate observation from the new state's emission distribution
        # hmm_model.means_ is (n_components, n_features)
        # hmm_model.covars_ is (n_components, n_features, n_features) for 'full'
        mean_vector = hmm_model.means_[next_state, :]
        cov_matrix = hmm_model.covars_[next_state, :, :]
        
        # Ensure covariance matrix is positive semi-definite for multivariate_normal
        # Add small identity matrix if necessary (jitter)
        try:
            next_obs_scaled = local_random_state.multivariate_normal(mean_vector, cov_matrix)
        except np.linalg.LinAlgError:
            print(f"Warning: Covariance matrix for state {next_state} not positive definite. Adding jitter.")
            jitter = 1e-6 * np.eye(len(mean_vector))
            next_obs_scaled = local_random_state.multivariate_normal(mean_vector, cov_matrix + jitter)

        sim_observations_scaled.append(next_obs_scaled)
        
        # Update current state and observation for next step
        current_state = next_state
        current_obs = next_obs_scaled # This line was missing in template's logic for iterative simulation
                                     # but the template actually fed last_observation repeatedly to model.predict,
                                     # which is not true simulation. This version is a proper simulation.
                                     # If you want to predict states based on *actual* future data (if available), that's different.

    return sim_states, np.array(sim_observations_scaled)


#Cell 25 (should be 26):Simulate future regimes
n_future_steps = 30  
last_observation_scaled = X_scaled[-1] # Use the last observation from the scaled dataset

# Simulate using the model trained on the FULL dataset (`model`)
future_sim_regimes, future_sim_observations_scaled = simulate_future_regimes(
    model, last_observation_scaled, n_future_steps
)

last_date = df['datetime'].iloc[-1]
# Assuming hourly data, generate future dates accordingly
future_dates = pd.date_range(start=last_date + timedelta(hours=1), periods=n_future_steps, freq='H')


future_df = pd.DataFrame({
    'datetime': future_dates,
    'regime_simulated': future_sim_regimes # Simulated regimes
})

# Inverse transform the simulated observations to their original scale
future_sim_observations_original = scaler.inverse_transform(future_sim_observations_scaled)
for i, feature_name in enumerate(features):
    future_df[feature_name + '_simulated'] = future_sim_observations_original[:, i]


#Cell 26 (should be 27):Visualize historical data and future SIMULATED regimes & observations
plt.figure(figsize=(16, 8))
plot_feature_sim_viz = features[0] # Feature to plot

# Plot Historical Data and True Regimes
plt.plot(df['datetime'], df[plot_feature_sim_viz], 'k-', alpha=0.3, label=f'Historical {plot_feature_sim_viz}')
for state in range(optimal_n_states):
    mask_hist = (df['regime'] == state) # Using 'regime' from full model fit
    if np.any(mask_hist):
        plt.scatter(df.loc[mask_hist, 'datetime'], 
                    df.loc[mask_hist, plot_feature_sim_viz], 
                    c=[colors(state)], # type: ignore
                    label=f'Historical Regime {state}',
                    alpha=0.5, s=30)

# Vertical line for current date (end of historical data)
plt.axvline(x=last_date, color='r', linestyle='--', label='Current Date (End of Historical)')

# Plot Simulated Future Observations
plt.plot(future_df['datetime'], future_df[plot_feature_sim_viz + '_simulated'], 'b--', 
         linewidth=2, label=f'Simulated Future {plot_feature_sim_viz}')

# Scatter plot for Simulated Future Regimes
for state in range(optimal_n_states):
    mask_future = (future_df['regime_simulated'] == state)
    if np.any(mask_future):
        plt.scatter(future_df.loc[mask_future, 'datetime'], 
                    future_df.loc[mask_future, plot_feature_sim_viz + '_simulated'], 
                    c=[colors(state)], # type: ignore
                    marker='x',
                    label=f'Simulated Future Regime {state}',
                    alpha=0.9, s=100)

plt.title(f'Historical Data & Future Simulated Regimes/Observations ({plot_feature_sim_viz})')
plt.xlabel('Date')
plt.ylabel(plot_feature_sim_viz)
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.grid(True)
plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()


#Cell 27 (should be 28):Calculate regime characteristics and interpretations
print("\nRegime interpretations (based on full data model):")
for state in range(optimal_n_states):
    if state not in regime_stats.index:
        print(f"\nRegime {state}: Not observed in data or insufficient data.")
        continue
        
    print(f"\nRegime {state}:")
    
    feature_means = [regime_stats.loc[state, (feature, 'mean')] for feature in features]
    feature_stds = [regime_stats.loc[state, (feature, 'std')] for feature in features]
    
    for i, feature in enumerate(features):
        mean_val = feature_means[i]
        std_val = feature_stds[i]
        
        all_means_for_feature = [regime_stats.loc[s, (feature, 'mean')] for s in regime_stats.index]
        # Check if all_means_for_feature is not empty before calculating median
        if all_means_for_feature:
            median_overall_mean = np.median(all_means_for_feature)
            level_desc = "High" if mean_val > median_overall_mean else "Low" if mean_val < median_overall_mean else "Average"
            print(f"- {level_desc} {feature}: {mean_val:.2f} (±{std_val:.2f})")
        else:
            print(f"- {feature}: {mean_val:.2f} (±{std_val:.2f}) (Comparison median not available)")

            
    if state in transition_probs.index and state in transition_probs.columns:
        stability = transition_probs.loc[state, state]
        stability_level = "High" if stability > 0.7 else "Moderate" if stability > 0.4 else "Low"
        print(f"- {stability_level} stability (self-transition probability: {stability:.2f})")
    else:
        print("- Stability: Data not available for this state in transition matrix.")
        
    state_durations_hrs = duration_df[duration_df['regime'] == state]['duration_hours']
    if not state_durations_hrs.empty:
        avg_duration_hrs = state_durations_hrs.mean()
        print(f"- Average duration: {avg_duration_hrs:.2f} hours")
    else:
        print("- Average duration: No occurrences of this regime.")
            
    frequency = (df['regime'] == state).mean() * 100
    print(f"- Frequency: {frequency:.2f}% of the time")

#Cell 28 (should be 29):Save the model
model_package = {
    'model': model, # Model trained on full data
    'scaler': scaler,
    'optimal_n_states': optimal_n_states,
    'features': features,
    'train_test_split_index': train_size, # Store the index for reference
    'creation_date': datetime.now().strftime('%Y-%m-%d %H:%M:%S')
}

model_filename = 'btc_gold_flow_hmm_model.pkl'
try:
    with open(model_filename, 'wb') as f:
        pickle.dump(model_package, f)
    print(f"\nModel successfully saved to '{model_filename}'")
except Exception as e:
    print(f"\nError saving model: {e}")

print("\nHMM analysis complete!")


# --- Advanced Validation Section (Cells 29-34 from original template) ---
# This section uses 'hidden_states' (from model on full data) as "true" labels for evaluation,
# and 'train_regimes'/'test_regimes' (from model on train data, predicted on train/test) as "predicted".
# This evaluates how well a model trained on a subset (train_data) can replicate the
# regime structure found by a model trained on the full dataset.

#Cell 29 (from template, now ~30): Regime Remapping & Accuracy Functions
from sklearn.metrics import adjusted_rand_score, normalized_mutual_info_score, adjusted_mutual_info_score
from scipy.optimize import linear_sum_assignment

def map_regimes_by_feature_means(true_regimes_series, pred_regimes_series, data_df_for_true, data_df_for_pred, feature_list):
    """Maps predicted regimes to true regimes based on similarity of feature means."""
    df_true_temp = data_df_for_true.copy()
    df_true_temp['regime_true'] = true_regimes_series
    
    df_pred_temp = data_df_for_pred.copy()
    df_pred_temp['regime_pred'] = pred_regimes_series

    true_regime_means = {}
    for reg_val in np.unique(true_regimes_series):
        if not df_true_temp[df_true_temp['regime_true'] == reg_val][feature_list].empty:
            true_regime_means[reg_val] = df_true_temp[df_true_temp['regime_true'] == reg_val][feature_list].mean().values
        else: # Handle cases where a regime might not be in the subset
             true_regime_means[reg_val] = np.full(len(feature_list), np.nan)


    pred_regime_means = {}
    for reg_val in np.unique(pred_regimes_series):
        if not df_pred_temp[df_pred_temp['regime_pred'] == reg_val][feature_list].empty:
            pred_regime_means[reg_val] = df_pred_temp[df_pred_temp['regime_pred'] == reg_val][feature_list].mean().values
        else:
            pred_regime_means[reg_val] = np.full(len(feature_list), np.nan)

    
    true_labels_unique = sorted(list(true_regime_means.keys()))
    pred_labels_unique = sorted(list(pred_regime_means.keys()))
    
    # Cost matrix: Euclidean distance between mean feature vectors of regimes
    cost_matrix = np.zeros((len(true_labels_unique), len(pred_labels_unique)))
    for i, true_label in enumerate(true_labels_unique):
        for j, pred_label in enumerate(pred_labels_unique):
            # Handle potential NaNs if a regime was empty for mean calculation
            if np.isnan(true_regime_means[true_label]).any() or np.isnan(pred_regime_means[pred_label]).any():
                cost_matrix[i, j] = np.inf # Assign high cost if means are undefined
            else:
                cost_matrix[i, j] = np.linalg.norm(true_regime_means[true_label] - pred_regime_means[pred_label])
    
    # Hungarian algorithm for optimal assignment
    true_indices, pred_indices = linear_sum_assignment(cost_matrix) # row_ind maps to true, col_ind to pred
    
    mapping = {} # pred_label -> true_label
    for i, j in zip(true_indices, pred_indices):
        mapping[pred_labels_unique[j]] = true_labels_unique[i]
        
    return mapping

def remap_predicted_regimes(pred_regimes_series, mapping_dict):
    return np.array([mapping_dict.get(pred_label, pred_label) for pred_label in pred_regimes_series])


# "True" regimes from the model trained on the full dataset
true_regimes_train_part = hidden_states[:train_size]
true_regimes_test_part = hidden_states[train_size:]

# Predicted regimes from the model trained only on X_train
# train_regimes = train_model.predict(X_train) -> already done
# test_regimes = train_model.predict(X_test) -> already done

# Calculate mapping based on training data portion
# We need the original unscaled data (df) to calculate feature means for mapping
# df_train and df_test are already defined
# df_train['regime_true_full_model'] = true_regimes_train_part
# df_train['regime_predicted_train_model'] = train_regimes

print("\n--- Advanced Validation: Comparing Train-Only Model to Full-Data Model ---")

# Map regimes from train_model (trained on X_train) to match regimes from model (trained on X_scaled)
# using the training portion of the data for establishing the map.
# Note: df_train contains original feature values.
# We need to pass the unscaled feature dataframes `df_train` and `df_test` for calculating means.
# `true_regimes_train_part` are labels on `df_train` from the full model.
# `train_regimes` are labels on `df_train` from the train_model.

mapping_train_to_full = map_regimes_by_feature_means(
    true_regimes_series=pd.Series(true_regimes_train_part, index=df_train.index), 
    pred_regimes_series=pd.Series(train_regimes, index=df_train.index),
    data_df_for_true=df_train, # Data corresponding to true_regimes_train_part
    data_df_for_pred=df_train, # Data corresponding to train_regimes
    feature_list=features
)
print(f"Regime mapping (Train-Model's states to Full-Model's states): {mapping_train_to_full}")

# Remap predicted regimes from train_model for both train and test sets using this mapping
remapped_train_regimes = remap_predicted_regimes(train_regimes, mapping_train_to_full)
remapped_test_regimes = remap_predicted_regimes(test_regimes, mapping_train_to_full)

# Calculate accuracy
train_accuracy_aligned = np.mean(true_regimes_train_part == remapped_train_regimes)
test_accuracy_aligned = np.mean(true_regimes_test_part == remapped_test_regimes)

print(f"Aligned Regime Prediction Accuracy (Train set): {train_accuracy_aligned:.4f}")
print(f"Aligned Regime Prediction Accuracy (Test set): {test_accuracy_aligned:.4f}")


#Cell 31 (from template): Clustering evaluation metrics
print("\nClustering Evaluation Metrics (Comparing Train-Model predictions to Full-Model states):")
# Training Data Part
ari_train = adjusted_rand_score(true_regimes_train_part, remapped_train_regimes)
nmi_train = normalized_mutual_info_score(true_regimes_train_part, remapped_train_regimes)
ami_train = adjusted_mutual_info_score(true_regimes_train_part, remapped_train_regimes)
print("Training Data:")
print(f"  Adjusted Rand Index (ARI): {ari_train:.4f}")
print(f"  Normalized Mutual Information (NMI): {nmi_train:.4f}")
print(f"  Adjusted Mutual Information (AMI): {ami_train:.4f}")

# Test Data Part
ari_test = adjusted_rand_score(true_regimes_test_part, remapped_test_regimes)
nmi_test = normalized_mutual_info_score(true_regimes_test_part, remapped_test_regimes)
ami_test = adjusted_mutual_info_score(true_regimes_test_part, remapped_test_regimes)
print("Test Data:")
print(f"  Adjusted Rand Index (ARI): {ari_test:.4f}")
print(f"  Normalized Mutual Information (NMI): {nmi_test:.4f}")
print(f"  Adjusted Mutual Information (AMI): {ami_test:.4f}")


#Cell 32 (from template): Visualize original, predicted (train_model), and remapped regimes
fig, axes = plt.subplots(3, 1, figsize=(16, 15), sharex=True, sharey=True)
plot_feature_val_viz = features[0]

# Plot 1: True Regimes (from model trained on full data)
axes[0].set_title(f'Full-Data Model Regimes ({plot_feature_val_viz})')
for state in range(optimal_n_states):
    # Train part
    mask_true_train = (true_regimes_train_part == state)
    if np.any(mask_true_train):
        axes[0].scatter(dates_train[mask_true_train], df_train.loc[mask_true_train, plot_feature_val_viz], 
                        c=[colors(state)], label=f'Regime {state}', alpha=0.7, s=20) # type: ignore
    # Test part
    mask_true_test = (true_regimes_test_part == state)
    if np.any(mask_true_test):
        axes[0].scatter(dates_test[mask_true_test], df_test.loc[mask_true_test, plot_feature_val_viz], 
                        c=[colors(state)], alpha=0.7, s=20, marker='o') # type: ignore
axes[0].axvline(dates_train.iloc[-1], color='r', linestyle='--', label='Train/Test Split')
axes[0].legend(loc='upper left', bbox_to_anchor=(1,1))
axes[0].set_ylabel(plot_feature_val_viz)

# Plot 2: Predicted Regimes (from train_model, BEFORE remapping)
axes[1].set_title(f'Train-Only Model Predicted Regimes (Before Remapping) ({plot_feature_val_viz})')
for state in range(optimal_n_states):
    # Train part
    mask_pred_train = (train_regimes == state)
    if np.any(mask_pred_train):
        axes[1].scatter(dates_train[mask_pred_train], df_train.loc[mask_pred_train, plot_feature_val_viz], 
                        c=[colors(state)], label=f'Pred Regime {state}', alpha=0.7, s=20) # type: ignore
    # Test part
    mask_pred_test = (test_regimes == state)
    if np.any(mask_pred_test):
        axes[1].scatter(dates_test[mask_pred_test], df_test.loc[mask_pred_test, plot_feature_val_viz], 
                        c=[colors(state)], alpha=0.7, s=20, marker='x') # type: ignore
axes[1].axvline(dates_train.iloc[-1], color='r', linestyle='--')
axes[1].legend(loc='upper left', bbox_to_anchor=(1,1))
axes[1].set_ylabel(plot_feature_val_viz)

# Plot 3: Remapped Predicted Regimes (from train_model, AFTER remapping)
axes[2].set_title(f'Train-Only Model Predicted Regimes (After Remapping to Full-Data Model States) ({plot_feature_val_viz})')
for state in range(optimal_n_states):
    # Train part
    mask_remap_train = (remapped_train_regimes == state)
    if np.any(mask_remap_train):
        axes[2].scatter(dates_train[mask_remap_train], df_train.loc[mask_remap_train, plot_feature_val_viz], 
                        c=[colors(state)], label=f'Remapped Regime {state}', alpha=0.7, s=20) # type: ignore
    # Test part
    mask_remap_test = (remapped_test_regimes == state)
    if np.any(mask_remap_test):
        axes[2].scatter(dates_test[mask_remap_test], df_test.loc[mask_remap_test, plot_feature_val_viz], 
                        c=[colors(state)], alpha=0.7, s=20, marker='D') # type: ignore
axes[2].axvline(dates_train.iloc[-1], color='r', linestyle='--')
axes[2].legend(loc='upper left', bbox_to_anchor=(1,1))
axes[2].set_xlabel('Date')
axes[2].set_ylabel(plot_feature_val_viz)

plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()


#Cell 33 (from template): Analyze transition accuracy (comparing remapped train_model predictions to full_model states)
def identify_transitions(regime_array):
    """Returns indices where regime transitions occur (index is for the start of the new regime)"""
    if len(regime_array) < 2:
        return np.array([])
    changes = np.diff(regime_array) != 0
    change_indices = np.where(changes)[0] + 1  # +1 because diff reduces length by 1
    return change_indices

# Transitions in "true" states (from full model)
true_transitions_train = identify_transitions(true_regimes_train_part)
true_transitions_test = identify_transitions(true_regimes_test_part)

# Transitions in remapped predicted states (from train_model)
pred_transitions_train_remapped = identify_transitions(remapped_train_regimes)
pred_transitions_test_remapped = identify_transitions(remapped_test_regimes)

def transition_detection_metrics(true_trans, pred_trans, tolerance_window=1, length_of_series=None):
    """Calculates precision, recall, F1 for transition detection.
    Tolerance window: number of time steps around a true transition to count a prediction as correct.
    Length of series: used if you want to account for true negatives (not typical for this).
    """
    if not length_of_series: # Default to length of true transitions if not given
        length_of_series = len(true_trans) # This isn't quite right for general TN, but useful for context

    TP = 0
    FP = len(pred_trans) # Start by assuming all predicted are false positives
    
    # For True Positives and False Negatives
    matched_pred_indices = set()
    for t_idx in true_trans:
        found_match = False
        for p_idx_pos, p_idx_val in enumerate(pred_trans):
            if p_idx_pos in matched_pred_indices: # Already matched this predicted transition
                continue
            if abs(t_idx - p_idx_val) <= tolerance_window:
                TP += 1
                matched_pred_indices.add(p_idx_pos)
                found_match = True
                break # Match found for this true transition
        # If no match found, this true transition is a False Negative (implicitly handled by FN = len(true_trans) - TP)

    FP = len(pred_trans) - TP # Correct False Positives
    FN = len(true_trans) - TP # False Negatives
    
    precision = TP / (TP + FP) if (TP + FP) > 0 else 0
    recall = TP / (TP + FN) if (TP + FN) > 0 else 0
    f1 = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0
    
    return precision, recall, f1

print("\nRegime Transition Detection Metrics (Remapped Train-Model vs. Full-Model):")
# Training data part
train_prec, train_rec, train_f1 = transition_detection_metrics(
    true_transitions_train, pred_transitions_train_remapped, tolerance_window=1, length_of_series=len(true_regimes_train_part)
)
print("Training Data Transitions:")
print(f"  Precision: {train_prec:.4f}, Recall: {train_rec:.4f}, F1 Score: {train_f1:.4f}")

# Test data part
test_prec, test_rec, test_f1 = transition_detection_metrics(
    true_transitions_test, pred_transitions_test_remapped, tolerance_window=1, length_of_series=len(true_regimes_test_part)
)
print("Test Data Transitions:")
print(f"  Precision: {test_prec:.4f}, Recall: {test_rec:.4f}, F1 Score: {test_f1:.4f}")


#Cell 34 (from template): Visualize transitions
plt.figure(figsize=(16, 8))
# Plot one of the features as a backdrop
plt.plot(df['datetime'], df[plot_feature_val_viz], 'k-', alpha=0.15, label=f'{plot_feature_val_viz} (Full Data)')
plt.axvline(dates_train.iloc[-1], color='grey', linestyle='--', lw=1, label='Train/Test Split')

# Plot true transitions (from full_model)
y_max = df[plot_feature_val_viz].max()
y_min = df[plot_feature_val_viz].min()

# True transitions on training part
for idx in true_transitions_train:
    if idx < len(dates_train): # ensure index is within bounds
        plt.vlines(dates_train.iloc[idx], ymin=y_min, ymax=y_max, color='blue', alpha=0.6, lw=1.5, 
                   label='True Transition (Full Model)' if 'True Transition (Full Model)' not in plt.gca().get_legend_handles_labels()[1] else None)

# True transitions on test part
for idx in true_transitions_test:
    # idx is relative to the start of test set, need to adjust for plotting on full datetime axis
    actual_idx_in_df = train_size + idx
    if actual_idx_in_df < len(df['datetime']):
        plt.vlines(df['datetime'].iloc[actual_idx_in_df], ymin=y_min, ymax=y_max, color='blue', alpha=0.6, lw=1.5)


# Plot remapped predicted transitions (from train_model)
# Predicted remapped transitions on training part
for idx in pred_transitions_train_remapped:
    if idx < len(dates_train):
        plt.vlines(dates_train.iloc[idx], ymin=y_min*0.95, ymax=y_max*0.95, # Slight offset for visibility
                   color='green', alpha=0.6, lw=1.5, linestyle=':',
                   label='Predicted Transition (Train Model, Remapped)' if 'Predicted Transition (Train Model, Remapped)' not in plt.gca().get_legend_handles_labels()[1] else None)

# Predicted remapped transitions on test part
for idx in pred_transitions_test_remapped:
    actual_idx_in_df = train_size + idx
    if actual_idx_in_df < len(df['datetime']):
         plt.vlines(df['datetime'].iloc[actual_idx_in_df], ymin=y_min*0.95, ymax=y_max*0.95,
                   color='green', alpha=0.6, lw=1.5, linestyle=':')


plt.title('Regime Transitions: Full-Model (Blue) vs. Remapped Train-Model (Green Dashed)')
plt.xlabel('Date')
plt.ylabel(plot_feature_val_viz)
plt.legend(loc='upper left', bbox_to_anchor=(1,1))
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout(rect=[0, 0, 0.8, 1])
plt.show()

# Note: The original script had a merge_csv_files function at the end.
# This is not directly part of the HMM analysis shown above, so I've omitted it here.
# If you need that functionality, you can add it back.
print("\nEnd of script.")
```
